```python
labels_education = ['High school diploma', 'GED', 'Some college', "Bachelor's degree", "Master's degree", 'Doctorate', 'PHD']
labels_message_freq = ['Never', 'Other', 'Several times a year', 'Monthly', 'Weekly', 'Daily']

# Check if 'Education_Level' and 'Message_Exchange_Frequency' columns exist in merged_data
if 'Education_Level' in merged_data.columns and 'Message_Exchange_Frequency' in merged_data.columns:
    # Calculate the frequency of each education level
    education_freq = merged_data['Education_Level'].value_counts().reset_index()
    education_freq.columns = ['Education_Level', 'Frequency']

    # Replace the index with the corresponding labels for education levels
    education_freq['Education_Level'] = education_freq['Education_Level'].replace(dict(zip(range(len(labels_education)), labels_education)))

    # Calculate the frequency of each message exchange frequency level
    message_freq = merged_data['Message_Exchange_Frequency'].value_counts().reset_index()
    message_freq.columns = ['Message_Exchange_Frequency', 'Frequency']

    # Replace the index with the corresponding labels for message exchange frequency levels
    message_freq['Message_Exchange_Frequency'] = message_freq['Message_Exchange_Frequency'].replace(dict(zip(range(len(labels_message_freq)), labels_message_freq)))

    # Find the predominant education level
    predominant_education_level = education_freq.loc[0, 'Education_Level']

    # Find the predominant message exchange frequency level
    predominant_message_freq = message_freq.loc[0, 'Message_Exchange_Frequency']

    # Print the predominant education level and message exchange frequency
    print(f"The predominant education level of your client's target customers is: {predominant_education_level}")
    print(f"The predominant message exchange frequency of your client's target customers is: {predominant_message_freq}")
```

```python
from pandas.plotting import parallel_coordinates

# Select relevant columns for parallel plot
parallel_data = merged_data[['Gender', 'Ethnicity', 'Annual_Income', 'Message_Exchange_Frequency', 
                             'Importance_of_Message_Exchange', 'Messaging_App_Usage_Frequency', 
                             'Message_Exchange_People_Count']]

# Replace gender labels using .loc to modify the DataFrame in place
parallel_data.loc[parallel_data['Gender'] == 1, 'Gender'] = 'Male'
parallel_data.loc[parallel_data['Gender'] == 2, 'Gender'] = 'Female'

# Create a copy of the DataFrame to avoid SettingWithCopyWarning
parallel_data = parallel_data.copy()

# Preprocess the 'Annual_Income' column using .loc to modify the DataFrame in place
income_map = {
    '$29,999 or less': 1,
    '$30,000 to $49,999': 2,
    '$50,000 to $69,999': 3,
    '$70,000 to $89,999': 4,
    '$90,000 to $91,999': 5,
    '$92,000 or more': 6
}
parallel_data['Annual_Income'] = parallel_data['Annual_Income'].map(income_map)

# Plot parallel coordinates
plt.figure(figsize=(16, 10))
parallel_coordinates(parallel_data, 'Gender', colormap='viridis')
plt.title('Message App Users Demographic Factors')
plt.legend(loc='best')
plt.tight_layout()
plt.savefig('parallel.png')
plt.show()
```

```python
import warnings

# Filter relevant columns
plot_data = merged_data[['Usage timeframe: 8am_to_9:59am', 'Radio_Listening_Frequency', 'Brand', 'Ethnicity']]

# Define the renaming dictionaries for Ethnicity and x/y-axis labels
ethnicity_labels = {
    1: 'American Indian',
    2: 'African American',
    3: 'Mexican American',
    4: 'Other Hispanic',
    5: 'Asian',
    6: 'Southeast Asian',
    7: 'Pacific Islander',
    8: 'Filipino',
    9: 'Multiple Ethnicities',
    10: 'White',
    11: 'Other/Not Stated',
    12: 'International'
}

x_axis_labels = {
    -3: 'Does not motivate',
    -2: 'Strongly disagree',
    -1: 'Disagree',
    0: 'Unsure',
    1: 'Agree',
    2: 'Strongly agree'
}

y_axis_labels = {
    0: 'Never',
    1: 'Less than 1 hour',
    2: '1 to almost 3 hours',
    3: '3 to almost 5 hours',
    4: '5 to almost 7 hours',
    5: '7 to almost 9 hours',
    6: '9 or more hours'
}

# Use .loc to map values without the warning
with warnings.catch_warnings():
    warnings.filterwarnings('ignore')
    plot_data.loc[:, 'Ethnicity'] = plot_data['Ethnicity'].map(ethnicity_labels)

# Create a swarm plot with renamed axis labels
plt.figure(figsize=(16, 10))
sns.swarmplot(
    x='Brand',
    y='Radio_Listening_Frequency',
    hue='Ethnicity',
    data=plot_data,
    palette='Set2',
    dodge=True,
    size=8,
)
plt.xlabel('Brand')
plt.ylabel('Radio Listening Frequency')
plt.title('Message App User Preferences  Radio Listening Frequency by Brand and Ethnicity')
plt.legend(title='Ethnicity', bbox_to_anchor=(1.05, 0.5), loc='center left', ncol=1)
plt.yticks(list(y_axis_labels.keys()), list(y_axis_labels.values()))  # Rename y-axis labels
plt.grid(True)
plt.tight_layout()

plt.savefig('swarm_plot_message_app.png')

plt.show()
```
